```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import json
import string
import re
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import os
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def preprocess_a(input):
    output = input.lower()
    # print(output)
    return preprocess_b(output)
def preprocess_b(input):
    words = input.split(' ')
    # words = [''.join(word.split(' ')) for word in words]
    result = ' '.join(words)
    tokens = nltk.word_tokenize(result)
    # print(tokens)
    return preprocess_c(tokens)
def preprocess_c(text):
    # for i in text:
    #     str+=
    output = ' '.join(text)
    # print(output)
    output = output.replace("'",'')
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(output)
    filtered_text = [word for word in word_tokens if word not in stop_words]
    # print( ' '.join(filtered_text))
    return preprocess_d( ' '.join(filtered_text))
def preprocess_d(input):
    words = input.split(' ')
    # words = [''.join(word.split(' ')) for word in words]
    result = ' '.join(words)
    result = result.translate(str.maketrans('', '', string.punctuation))
    tokens = nltk.word_tokenize(result)
    # print(tokens)
    return preprocess_e(tokens)
def preprocess_e(text):
    text = ' '.join(text)
    output = text.replace("'", "")
    word_tokens = word_tokenize(output)
    stop_words = set(stopwords.words('english'))
    filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
    porter = PorterStemmer()
    stemmed_text = [porter.stem(word) for word in filtered_text]
    lemmatizer = WordNetLemmatizer()
    lemmatized_text = [lemmatizer.lemmatize(word) for word in stemmed_text]
    preprocessed_text = ' '.join(lemmatized_text)
    
    return preprocessed_text
```
